```python
# Auth
client_id = dbutils.secrets.get(scope="petiot", key="client-id")
print("Secret scope OK ✓")

# Storage
storage_account = "petiot"
spark.conf.set(f"fs.azure.account.auth.type.{storage_account}.dfs.core.windows.net", "OAuth")
spark.conf.set(f"fs.azure.account.oauth.provider.type.{storage_account}.dfs.core.windows.net",
               "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set(f"fs.azure.account.oauth2.client.id.{storage_account}.dfs.core.windows.net",
               dbutils.secrets.get(scope="petiot", key="client-id"))
spark.conf.set(f"fs.azure.account.oauth2.client.secret.{storage_account}.dfs.core.windows.net",
               dbutils.secrets.get(scope="petiot", key="client-secret"))
spark.conf.set(f"fs.azure.account.oauth2.client.endpoint.{storage_account}.dfs.core.windows.net",
               f"https://login.microsoftonline.com/{dbutils.secrets.get('petiot','tenant-id')}/oauth2/token")

for container in ["raw", "bronze", "checkpoints"]:
    items = dbutils.fs.ls(f"abfss://{container}@{storage_account}.dfs.core.windows.net/")
    print(f"Container '{container}' accessible ✓  ({len(items)} items)")
```
