# Reading a file from json

```python
import gzip

raw_path = "abfss://raw@petiot.dfs.core.windows.net/health_snapshots/year=2025/month=01/day=03/health_snapshots.json"

content = spark.read.format("binaryFile").load(raw_path).select("content").first()[0]
json_lines = gzip.decompress(bytes(content)).decode("utf-8").splitlines()
health_snapshots_df = spark.read.json(spark.sparkContext.parallelize(json_lines))

display(health_snapshots_df.limit(5))
```
