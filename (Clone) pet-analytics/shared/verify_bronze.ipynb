```python
storage_account = "petiot"

tables = [
    "owners", "pets", "products", "firmware", "batches", "devices",
    "sales", "device_snapshots", "activity_snapshots", "health_snapshots",
    "feeding_events", "hydration_events", "device_failures", "firmware_deployments"
]

print(f"{'Table':<30}  {'Rows':>10}  {'Columns':>10}  {'Status'}")
print("─" * 65)

for table in tables:
    path = f"abfss://bronze@{storage_account}.dfs.core.windows.net/{table}"
    try:
        df    = spark.read.format("delta").load(path)
        rows  = df.count()
        cols  = len(df.columns)
        print(f"  {table:<28}  {rows:>10,}  {cols:>10}  ✓")
    except Exception as e:
        print(f"  {table:<28}  {'ERROR':>10}  {'─':>10}  ✗  {str(e)[:40]}")
```
